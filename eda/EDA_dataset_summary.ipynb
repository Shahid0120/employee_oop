# Exploratory Data Analysis : Finding Dtype of Values 
In this notebook the main goals:
- Identify all datatypes from dataset files
- Identify columns data ranges
- Identify if missing columns

## Requirements for print function
Automate column and structure for Markdown file in format 

If object has more then 30 in list then dont show all list but 1 object of list
- PayZone (object) : ['Zone C' 'Zone A' 'Zone B']
- Division (object) : Finance & Accounting

Else if dtype is object but a date then
- StartDate (object) : Day-Month-Year / 20-Sep-19


If int64 
- EmpID (int64) : XXXX / 3942

```python
import pandas as pd
import random 

# List of DataFrame names as strings
df_list = [
    'employee_engagement_survey_data_df', 
    'employee_data_df', 
    'recruitment_data_df', 
    'training_and_development_data_df'
]

# Dictionary to hold DataFrames
dataframes = {}

print("Printing all columns in Above Requirements...")
try: 
    print("Importing all Dataset as DataFrame...")
    # Read CSV files into DataFrames and store them in the dictionary
    dataframes['employee_engagement_survey_data_df'] = pd.read_csv("../data/employee_engagement_survey_data.csv")
    dataframes['employee_data_df'] = pd.read_csv('../data/employee_data.csv')
    dataframes['recruitment_data_df'] = pd.read_csv('../data/recruitment_data.csv')
    dataframes['training_and_development_data_df'] = pd.read_csv('../data/training_and_development_data.csv')
    print("Finished Importing Data as DataFrame...")
except Exception as e:
    print(f"Failed to import data: {e}")

print("\n")

# Iterate through each DataFrame in the dictionary
for df_name in df_list:
    df = dataframes[df_name]
    print(f"Starting Columns in {df_name}....")
    # Iterate through columns in the DataFrame
    for col in df.columns:
        
        # Generated Random Index
        random_idx = random.randint(0, len(df[col]) - 1) 

        # for int64 and object function with unique list < 15
        if df[col].dtype == 'int64' or len(df[col].unique()) > 15: 
            print(f"- {col} | dtype: {df[col].dtype} | {df.loc[random_idx, col]}")
        else:
            # all other object functions
            if len(df[col].unique()) < 15:
                 print(f"- {col} | dtype: {df[col].dtype} | {df[col].unique()}")

    print(f"Finished printing all Columns in {df_name}...")
    print(f"\n")
```
